```python
# specifying directory containing files 
directory_path="KO-KEGG\ortologs\\"

file_pattern = '*.tsv'
file_paths = glob.glob(directory_path + file_pattern)

print(file_paths)
```

```python
def pond_ex(path):
    l=path.split('\\')
    fn=l[-1]
    fn=fn.replace('_pred_metagenome_unstrat.tsv','')
    return fn
pond_ex('KO-KEGG\\ortologs\\S08_pred_metagenome_unstrat.tsv')   
```

```python
#creating initial dataframe 
result_df=pd.read_csv(file_paths[0],sep="\t",header=None)
result_df=result_df.T
colnames=result_df.iloc[0]
result_df.columns = colnames
result_df = result_df.drop(result_df.index[0])

result_df.insert(0,'pond',pond_ex(file_paths[0]))
#result_df = result_df.drop(["function"],axis = 1)
result_df.head()
```

```python
# Iterate over the file paths and read each CSV file

for file_path in file_paths[1:]:
    print("reading",file_path)
    df = pd.read_csv(file_path,header=None, sep ="\t")
    print("transforming")
    df=df.T
    
    colnames=df.iloc[0]
    df.columns = colnames
    df = df.drop(df.index[0])
    df.insert(0,'pond',pond_ex(file_path))
    print("merging")
    result_df = pd.merge(result_df, df, how='outer')


result_df = result_df.drop(["function"],axis = 1)
print(result_df.shape)

result_df.head(100)
```

```python
result_df.to_csv('ortologs_full_table', index=False)
```
