```python
# specifying directory containing files 
directory_path="H2O_KO_unmerged/"

file_pattern = '*.tsv'
file_paths = glob.glob(directory_path + file_pattern)

print(file_paths)
```

```python
def pond_ex(path):
    l=path.split('\\')
    fn=l[-1]
    fn_l = fn.split('_')
    
    fn=fn_l[0]
    return fn
pond_ex('H2O_KO_unmerged\\S08_pred_metagenome_unstrat.tsv')   
```

```python
#creating initial dataframe 
result_df=pd.read_csv(file_paths[0],header=None,sep='\t')
result_df=result_df.T
colnames=result_df.iloc[0]
result_df.columns = colnames
result_df = result_df.drop(result_df.index[0])
sample_id = result_df['function'].str.split('.',expand=True)[1]
result_df.insert(0,'pond',pond_ex(file_paths[0]))
result_df.insert(1,'sample_id',sample_id)



result_df.head()
```

```python
# Iterate over the file paths and read each CSV file

for file_path in file_paths[1:2]:
    
    df = pd.read_csv(file_path,header=None,sep='\t')
    df=df.T
    
    colnames=df.iloc[0]
    df.columns = colnames
    df = df.drop(df.index[0])
    df.insert(0,'pond',pond_ex(file_path))
    sample_id = result_df['function'].str.split('.',expand=True)[1]
    result_df['sample_id'] = sample_id

    result_df = pd.merge(result_df, df, how='outer')
    print("Merging",file_path)


print(result_df.shape)

result_df.head(100)
```
