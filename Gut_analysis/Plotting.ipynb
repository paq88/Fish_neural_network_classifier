```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('Fish_results.csv')
```

```python
#df = df.iloc[1:,:]
```

```python
class_5_df = df[df['classes'] == 5]
```

```python
validation_means = class_5_df["val_acc_mean"] 
```

```python
u_green = '#004651'
u_red = '#742732' 
u_grey = '#474747'
```

```python

# Data: Replace these with your actual values
approaches = ['X - FNN', 'X - CNN', 'emb - FNN','emb - CNN','PCA - FNN','PCA - CNN']  # Names of the approaches
validation_means = class_5_df["val_acc_mean"]  # Validation mean accuracies
validation_sds = class_5_df["val_acc_sd"]    # Standard deviations for validation accuracies
test_accuracies = class_5_df["test_acc"]  # Test accuracies

x = np.arange(len(approaches))


# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot validation accuracies with error bars (lower zorder)
ax.errorbar(x, validation_means, yerr=validation_sds, fmt='o', capsize=5, label='Validation Accuracy', color='black', markersize=8, alpha=1, zorder=1)

# Plot test accuracies as points (higher zorder)
ax.scatter(x, test_accuracies, label='Test Accuracy', color='r', s=80, marker='x', zorder=2)

# Add labels, title, and legend
ax.set_xlabel('Approaches')
ax.set_ylabel('Accuracy')
ax.set_title('5 Class Classification')
ax.set_xticks(x)
ax.set_xticklabels(approaches)
ax.set_ylim(0, 1)

# Add horizontal line for baseline (set zorder appropriately)
ax.axhline(y=0.56, color=u_green, linestyle='--', label='baseline (xgboost)', linewidth=1, alpha=1, zorder=0)
ax.axhline(y=0.2, color=u_red, linestyle='--', label='random', linewidth=1, alpha=1, zorder=0)

# Add legend
ax.legend()

# Show grid for better readability
ax.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```
